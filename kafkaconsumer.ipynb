```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer(
    'demm',
     bootstrap_servers=['3.149.239.245:9092'],
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
# Initialize S3FileSystem with the correct AWS region
s3 = s3fs.S3FileSystem(client_kwargs={'region_name': 'us-east-2'})

# Consume messages from Kafka and upload each one to S3 with a unique file name
for count, message in enumerate(consumer):
    # Construct a unique file name for each message
    s3_file_path = f"s3://awais9010-stock-marketproject/stock_market_{count}.json"
    # Write the JSON message to S3
    with s3.open(s3_file_path, 'w') as f:
        json.dump(message.value, f)
    print(f"Uploaded message {count} to {s3_file_path}")
```
